```python
import os 
  
# Function to rename multiple files 
def main(): 
  
    for count, filename in enumerate(os.listdir(r"C:\Users\ajarun\ineuron\dataAugmentation\new")): 
        dst ="nnmask" + str(count) + ".jpeg"
        src = filename 
        dst = dst 
        os.chdir(r"C:\Users\ajarun\ineuron\dataAugmentation\new")
        # rename() function will 
        # rename all the files 
        os.rename(src, dst) 
  
# Driver Code 
if __name__ == '__main__': 
      
    # Calling main() function 
    main() 
```
